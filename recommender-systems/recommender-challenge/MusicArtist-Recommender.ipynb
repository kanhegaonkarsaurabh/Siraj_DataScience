```python
#Dependencies required
from fetch_lastfm import fetch_lastfm
from lightfm import LightFM
import numpy as np
```

```python
# Lastfm's first 100k user's dataset
data = fetch_lastfm()
data
```

```python
# Define a LightFM WARP based model which works on the data matrix i.e. the training dataset
model = LightFM(loss='warp')
model.fit(data['matrix'],epochs=30,num_threads=1)
```

```python
# This is a recommendation function which takes the model and the dataset and gives a recommendation on the basis of user
def sample_recommendation(model,coo_matrix,user_ids):
    n_items = coo_matrix.shape[1] # here, n_items is basically the no. of columns
    
    for user in user_ids:
        scores = model.predict(user,np.arange(n_items))
        top_scores = np.argsort(-scores)[:3]  # take the first three top_scores for the three users
        
        print('Recommendation for user %s' % user)
        
        # Iterate through the artists' dictionary and print the value as per the value id
        for x in top_scores.tolist():
            for artist,value in data['artists'].iteritems():
                if int(x) == value['id']: 
                    print("  - %s" % value['name'])
                    
    print '\n'    
        
```

```python
user_1 = raw_input("Please enter 1st user id")
user_2 = raw_input("Please enter 2nd user id")
user_3 = raw_input("Please enter 3rd user id")
sample_recommendation(model,data['matrix'],[user_1,user_2,user_3])
```
